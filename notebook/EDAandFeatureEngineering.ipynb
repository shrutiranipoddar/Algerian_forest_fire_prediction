## Data Cleaning

```python
dataset.loc[:122,"Region"] = 0
dataset.loc[122:,"Region"] = 1
df = dataset
```

```python
df[['month','day','year','Temperature','RH','Ws']] = df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
objects = [features for features in df.columns if df[features].dtype == 'object']
```

```python
percentage = df_copy['Classes'].value_counts(normalize = True)*100
```

```python
import matplotlib.pyplot as plt

# Get percentage and corresponding class labels
percentage = df_copy['Classes'].value_counts(normalize=True) * 100
label_map = {1: "Fire", 0: "Not Fire"}
classlabel = [label_map[i] for i in percentage.index]

# Plot
plt.figure(figsize=(12, 7))
plt.pie(percentage, labels=classlabel, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart of Classes")
plt.axis('equal')
plt.show()
```
